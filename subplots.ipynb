```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2 *x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(5,6),dpi=100)
axes[0][0].plot(a,b)
axes[0][0].set_title('TITLE 0,0')
axes[0][0].set_xlabel('X LABEL 0,0')
axes[0][0].set_ylabel('Y LABEL 0,0')
axes[0][0].set_xlim(0,10)
axes[0][0].set_ylim(0,10000)

axes[0][1].plot(x,y)
axes[0][1].set_title('TITLE 0,1')
axes[0][1].set_xlabel('X LABEL 0,1')
axes[0][1].set_ylabel('Y LABEL 0,1')
axes[1][0].set_xlim(0,10)
axes[1][0].set_ylim(0,20)

axes[1][0].plot(x,y)
axes[1][0].set_title('TITLE 1,0')
axes[1][0].set_xlabel('X LABEL 1,0')
axes[1][0].set_ylabel('Y LABEL 1,0')
axes[1][0].set_xlim(0,10)
axes[1][0].set_ylim(0,20)

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')
axes[1][1].set_xlabel('X LABEL 1,1')
axes[1][1].set_ylabel('Y LABEL 1,1')
axes[1][0].set_xlim(0,10)
axes[1][0].set_ylim(0,10000)

#for ax in axes:
    #ax.plot(x,y)

#axes[0].plot(x,y)

#plt.tight_layout()
#fig.subplots_adjust(wspace=0.5,hspace=0.5)
fig.suptitle('FIGURE LEVEL',fontsize=16,fontweight=50)
#fig.set_figwidth(10)
plt.tight_layout()
fig.savefig('record_subplots.png',bbox_inches='tight')
```
